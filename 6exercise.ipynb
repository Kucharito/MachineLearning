```python
X = df_scaled.values
```

```python

clustering_scores = []
for k in range(2, 11):
    kmeans = sklearn.cluster.KMeans(n_clusters=k, random_state=13)
    kmeans.fit(X)
    clustering_scores.append({
        'k': k,
        'sse': kmeans.inertia_,
        'silhouette': sklearn.metrics.silhouette_score(X, kmeans.labels_)
    })

df_clustering_scores=pd.DataFrame.from_dict(clustering_scores,orient='columns')
df_clustering_scores = df_clustering_scores.set_index('k')

```

```python
sns.lineplot(data=df_clustering_scores, y='silhouette', x='k')
```

```python
clustering=sklearn.cluster.KMeans(n_clusters=4, random_state=13)
clustering.fit(X)
```

```python
pd.Series(clustering.labels_).value_counts()
```

```python
df['cluster_id']=clustering.labels_
df.head()
```

```python
for column in df.select_dtypes(include=[np.number]).columns:
    if column != 'cluster_id':
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=df, x='cluster_id', y=column)
        plt.title(f"Boxplot of '{column}' by Cluster")
        plt.xlabel("Cluster ID")
        plt.ylabel(column)
        plt.show()
```

```python

df_only_numeric = df.select_dtypes(include=[np.number])

cluster_means = df.groupby('cluster_id')[df_only_numeric.columns].mean()

n_cols = len(df_only_numeric.columns)
fig, axes = plt.subplots(nrows=1, ncols=n_cols, figsize=(40, 6))


for i, column in enumerate(df_only_numeric.columns):
    sns.barplot(x=cluster_means.index, y=cluster_means[column], ax=axes[i]).set_title(column)
    axes[i].set_xlabel('Cluster ID')
    axes[i].set_ylabel('Mean Value')

plt.tight_layout()
plt.show()
```
